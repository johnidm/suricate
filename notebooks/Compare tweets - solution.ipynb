```python
tuites = [
    'Assim que você passar na padaria.',
    'Seu senso de humor é muito peculiar.',
    'Você é um rapaz de bom senso.',
    'Nessa área, podemos caçar legalmente alguns animais.',
    'Esse móvel precisa de um conserto urgente.',
    'Você é um rapaz de bom humor.',
    'Asso as batatas no forno por 40 minutos.',
    'Seu senso de direção é muito peculiar.',
    'Traz pão assim que você passar na padaria, por favor.',
    'Asso as batatas na padaria.',
]
```

```python
# return a tuple: text one, text two, similarity? (yes, no)
from difflib import SequenceMatcher

def similar(a, b):
    ratio = SequenceMatcher(None, a, b).ratio()
    return ratio
    
def check_similarity(a, b):
    return a, b, similar(a, b)
```

```python
import itertools

similarity = []

for a, b in itertools.combinations(tuites, 2):
    similarity.append(check_similarity(a, b))

import pprint 
pprint.pprint(similarity)
```

```python
[ (a, b) for a, b, ratio in similarity if ratio >= 0.7 ]
```
