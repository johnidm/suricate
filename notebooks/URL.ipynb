```python
import pymongo

MONGO_DB_URL='mongodb://localhost:27017/suricate'

client = pymongo.MongoClient(MONGO_DB_URL)
collection = client.suricate['bolsonaro']

data = list(collection.find({}))
```

```python
urls_raw = list(filter(None, [d['entities']['urls'] for d in data]))

from functools import reduce
import operator

url = [d['expanded_url'] for d in reduce(operator.concat, urls_raw)]

import pprint
pprint.pprint(url)
```
